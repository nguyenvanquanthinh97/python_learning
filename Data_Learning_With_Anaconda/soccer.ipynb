```python
# Rows with at least one NaN value
data_frame[pd.isnull(data_frame).any(axis=1)].head()
```

```python
# Drop rows with Nan values
data_frame = data_frame.dropna() # removes rows with any NaN values
data_frame = data_frame.reset_index() # reset's row indexes in case any rows were dropped
data_frame.head()
```

```python
# Dropping multiple columns
data_frame = data_frame.drop(columns=["sofifa_id", "long_name", "player_url"]) # we won't use text features for our initial basic models
data_frame.head()
```

```python
# Map feature values
data_frame["age"] = data_frame["age"].map(lambda age: "about to expire" if age > 32 else "still hired")
data_frame.head()
```

```python
# Lambda expressions to create new features
def get_BMI(height_cm, weight_kg):
    return round(weight_kg / ((height_cm / 100) * 2),1)
data_frame["bmi"] = data_frame[["height_cm", "weight_kg"]].apply(lambda row: get_BMI(row["height_cm"], row["weight_kg"]), axis=1)
data_frame.head()
```

```python
# Reorganize headers
data_frame = data_frame[["short_name", "long_name", "nationality","club"]]
data_frame.head()
```

```python
# Saving data to CSV
data_frame.to_csv("minimize_data.csv", index=False)
```
